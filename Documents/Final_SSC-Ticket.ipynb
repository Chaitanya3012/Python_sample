```python
#df['MTTR']=df['Actual Finish Adjusted']-df['Created Date Adjusted']
#df['MTTP']=df['Actual Start Adjusted']-df['Created Date Adjusted']
df['MTTR']=abs(df['Actual Finish Adjusted']-df['Created Date Adjusted']).astype('timedelta64[s]')/60
df['MTTP']=abs(df['Actual Start Adjusted']-df['Created Date Adjusted']).astype('timedelta64[s]')/60
```

```python
Mod_owner_group=[]

for group in Owner_Group:
    if group in "IBM-SMM-OPS":
        out = 'IBM-SMM-OPS'
    else:
        out = 'Others'
        
    Mod_owner_group.append(out)
```

```python
df1 = df[df['Console/Batch'] == 'Batch']
df1
```

```python
df2 = df[df['Console/Batch'] == 'Console']
df2
```

```python
data1 = df1[df1['Mod_group'] == 'IBM-SMM-OPS']
data2 = df1[df1['Mod_group'] == 'Others']
```

```python
data3 = df2[df2['Mod_group'] == 'IBM-SMM-OPS']
data4 = df2[df2['Mod_group'] == 'Others']
```

```python
print("MTTR(in mins) for Batch (IBM-SMM-OPS)group ")

a=data1.pivot_table(values='MTTR',index='Internal Priority',columns='Mod_group',aggfunc='count')
b=data1.pivot_table(values=['MTTR'],index='Internal Priority',columns='Mod_group',aggfunc='sum')
pd.concat((a,b),axis=1)
```

```python
print("MTTP(in mins) for Batch (Others)group")

c=data2.pivot_table(values='MTTP',index='Internal Priority',columns='Mod_group',aggfunc='count')
d=data2.pivot_table(values=['MTTP'],index='Internal Priority',columns='Mod_group',aggfunc='sum')
pd.concat((c,d),axis=1)
```

```python
print("MTTR(in mins) for Console (IBM-SMM-OPS)group")

e=data3.pivot_table(values='MTTR',index='Internal Priority',columns='Mod_group',aggfunc='count')
f=data3.pivot_table(values=['MTTR'],index='Internal Priority',columns='Mod_group',aggfunc='sum')
pd.concat((e,f),axis=1)
```

```python
print("MTTP(in mins) for Console(Others)group")

g=data4.pivot_table(values='MTTP',index='Internal Priority',columns='Mod_group',aggfunc='count')
h=data4.pivot_table(values=['MTTP'],index='Internal Priority',columns='Mod_group',aggfunc='sum')
pd.concat((g,h),axis=1)
```
